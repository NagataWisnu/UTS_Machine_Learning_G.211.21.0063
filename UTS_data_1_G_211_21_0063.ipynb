```python
# codingan untuk transformation data
from sklearn.preprocessing import LabelEncoder

enc = LabelEncoder()

dataset['Sex'] = enc.fit_transform(dataset['Sex'].values)
dataset['Obesity'] = enc.fit_transform(dataset['Obesity'].values)
dataset['DLP'] = enc.fit_transform(dataset['DLP'].values)
dataset['WPE'] = enc.fit_transform(dataset['WPE'].values)
dataset['Atypical'] = enc.fit_transform(dataset['Atypical'].values)
dataset['Nonanginal'] = enc.fit_transform(dataset['Nonanginal'].values)
dataset['LVH'] = enc.fit_transform(dataset['LVH'].values)
dataset['PRP'] = enc.fit_transform(dataset['PRP'].values)
dataset['CAD'] = enc.fit_transform(dataset['CAD'].values)
```

Hasil informasi data yang sudah diedit dan ambil data yang dibutuhkan

```python
dataset.info()
```

```python
dataset.head(10)
```

```python
# Codingan untuk membagi data X dan Y
x = dataset.iloc[:, 0:-1].values #Codingan untuk memasukkan dari data Age sampai BUN ke variabel x
y = dataset.iloc[:, -1].values   #Codingan untuk memasukkan dari data CAD ke variabel y
print(x)
print(y)
```

```python
x = dataset['CAD']
print(x)
```

```python
 # COdingan untuk membagi data Training dan Testing menggunakan library dari sklearn dengan function train_test_split
from sklearn.model_selection import train_test_split    #Codingan untuk mengimport library sklearn dan menggunakan function train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.25, random_state=0)    #Codingan untuk membagi data training menjadi 75% dan data testing menjadi (0.25) 25%
```

```python
# Codingan untuk memproses data dengan StandardScaler
from sklearn.preprocessing import StandardScaler    #Codingan untuk mengimport library sklearn dan menggunakan function StandardScaler
sc = StandardScaler()   #COdingan untuk memasukkan fungsi StandardScaler kedalam variabel sc

x_train = sc.fit_transform(x_train)   #Codingan Scaling untuk x_train
x_test = sc.fit_transform(x_test)     #Codingan Scaling skala untuk x_test
```

Memakai Algoritma K-nearest Neighbor (KNN)

```python
from sklearn.neighbors import KNeighborsClassifier
classifire = KNeighborsClassifier(n_neighbors=5, metric = 'minkowski', p=2)   #Codingan untuk menentukan k=5 (n_neighbors=5) dengan perhitungan jarak menggunakan metode minkowski
classifire.fit(x_train, y_train)    #Codingan untuk mengklasifikasikan data x_train dan y_train
```

```python
# Codingan untuk menentukan prediksi
y_pred = classifire.predict(x_test)   #Codingan untuk menentukan nilai prediksi pada label/target dengan variabel y_pred, dan data x_test digunakan untuk memprediksi
```

```python
# Mendapatkan nilai akurasi dengan Evaluasi dan Validasi
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)   #Membandingkan nilai label/target sebenarnya (y_test) dengan target prediksi (y_pred)
print(cm)   #[[49  5]     49 + 11 : jumlah data(76) x 100
            # [ 11 11]]   60 : 76 x  100 = 78,9% (Nilai Akurasi)
```

```python
print("Nilai akurasi : ", accuracy_score(y_test, y_pred))
```
